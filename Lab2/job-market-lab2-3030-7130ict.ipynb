```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market
df = pd.read_csv('job-market.csv')
print("The Dataset consists of: ", len(df), " rows and ", df.shape[1], " columns")
df.head(3)
```

```python
%matplotlib inline

plt.figure(figsize=(15,15))
ax=sns.countplot(y=df['Classification'],data=df,order=df['Classification'].value_counts().index)
ax.tick_params(labelsize=12) #axes x,y size by default
plt.title('Jobs and Number of Postings',fontsize=14)
plt.ylabel("Classification")
plt.xlabel("Number of Postings")  
```

```python
import matplotlib
#create dataframe for salary ranges
#df_salary=df[["LowestSalary","HighestSalary"]].value_counts().sort_index()
SalaryDf = df[df['LowestSalary'].isna()==False]
SalaryDf = SalaryDf.groupby('LowestSalary').count()
SalaryDf['Label'] = SalaryDf.index
SalaryDf['Count'] = SalaryDf["Id"]
SalaryDf.head()
```

```python
#salary ranges plot

# Pie Chart
plt.pie(SalaryDf['Count'], labels=SalaryDf['Label'],autopct='%.2f%%'
,radius=1.8,colors=['#b7e1a1','#e6f2a2','#ffe5ad','#ff9a8a']) #light grey green, peachy pink

# Adding Title of chart
plt.title("Job Posts by Salary Range",fontsize=12, pad=100)

# draw circle
centre_circle = plt.Circle((0, 0), 0.8, fc='white')
fig = plt.gcf()

# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
```
